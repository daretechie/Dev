```python
x = np.linspace(0, 5, 200)
y1 = 2 * x
y2 = x ** 2
y3 = np.log(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
x = np.linspace(0, 50, 100)
y = np.sin(x)
plt.figure( "MyFigure" )
plt.title("Sine Function")
plt.suptitle("Data Science")
plt.grid(True)
plt.plot(x, y)
plt.xlabel( "x-values" )
plt.ylabel( "y-values" )
plt.show()
```

```python
x = np.linspace( 10 , 50 , 100 )
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.log(x/ 3 )
plt.plot(x,y1, 'b-' , label = "Sine" )
plt.plot(x,y2, 'r-' , label = "Cosine" )
plt.plot(x,y3, 'g-' , label = "Logarithm" )
plt.legend( loc = 'upper left' )
plt.show()
```

## MATPLOTLIB PLOT TYPES

### HISTOGRAMS

```python
mu, sigma = 172 , 4

x = mu + sigma * np.random.randn( 10000 )
plt.xlabel( "Height" )
plt.ylabel( "Probability" )
plt.title( "Height of Students" )
plt.text( 160 , 0.125 ,"µ = 172, σ = 4" )
plt.axis([ 155 , 190 , 0 , 0.15 ])
plt.grid( True )
plt.hist(x, 100 , density = True , facecolor = "blue" )
```

### BAR CHART

```python
bob = ( 90 , 67 , 87 , 76 )
charles = ( 80 , 80 , 47 , 66 )
daniel = ( 40 , 95 , 76 , 89 )
skills = ( "Python" , "Java" , "Networking" , "Machine Learning" )

width = 0.2
index = np.arange( 4 )

plt.bar(index, bob, width =width, label = "Bob" )
plt.bar(index + width, charles, width =width, label = "Charles" )
plt.bar(index + width * 2 , daniel, width =width, label = "Daniel" )

plt.xticks(index + width, skills)
plt.ylim( 0 , 120 )
plt.title( "IT Skill Levels" )
plt.ylabel( "Skill Level" )
plt.xlabel( "IT Skill" )
plt.legend()
```

### PIE CHART

```python
labels = ( 'American' , 'German' , 'French' , 'Other' )
values = ( 47 , 23 , 20 , 10 )
plt.pie(values, labels =labels, autopct = "%.2f%%" , shadow = True )
plt.title( "Student Nationalities" )
plt.show()
```

### BOXPLOT

```python
mu, sigma = 172 , 4
values = np.random.normal(mu,sigma, 200 )
plt.boxplot(values)
plt.title( "Student's Height" )
plt.ylabel( "Height" )
plt.show()
```

```python
z = np.linspace( 0 , 20 , 100 )
x = np.sin(z)
y = np.cos(z)
ax = plt.axes( projection = '3d' )
ax.plot3D(x,y,z)
plt.show()
```

### SURFACE PLOTS

```python
def z_function(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2 ))
```

```python
x = np.linspace(- 5 , 5 , 50 )
y = np.linspace(- 5 , 5 , 50 )
X, Y = np.meshgrid(x,y)
Z = z_function(X,Y)
ax = plt.axes( projection = '3d' )
ax.plot_surface(X,Y,Z)
plt.show()
```
